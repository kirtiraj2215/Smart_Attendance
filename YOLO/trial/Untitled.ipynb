```python
from FaceAnalyzer import FaceAnalyzer
import cv2

# Initialize the FaceAnalyzer
fa = FaceAnalyzer()

# Load an image using OpenCV
image_path = "gp2.jpg"
frame = cv2.imread(image_path)

# Check if the image was loaded successfully
if frame is None:
    print("Error: Image not found or unable to load.")
    exit()

# Resize the image for easier processing (optional)
desired_width = 800
scale = desired_width / frame.shape[1]
new_dimensions = (desired_width, int(frame.shape[0] * scale))
frame_resized = cv2.resize(frame, new_dimensions, interpolation=cv2.INTER_AREA)

# Display the resized image
cv2.imshow("Resized Image", frame_resized)

# Convert the resized image to RGB (FaceAnalyzer expects RGB)
frame_rgb = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2RGB)

# Process the image with FaceAnalyzer
fa.process(frame_rgb)

# Check if faces are detected
if fa.nb_faces > 0:
    print(f"{fa.nb_faces} Faces found")

    for idx, face in enumerate(fa.faces):
        # Extract the face bounding box
        face_image = face.getFaceBox(frame_rgb)
        if face_image is not None:
            cv2.imshow(f"Face {idx + 1}", cv2.cvtColor(face_image, cv2.COLOR_RGB2BGR))  # Display in BGR

        # Extract the forehead region
        forehead_image = face.getFaceBox(frame_rgb, face.face_forhead_indices)
        if forehead_image is not None:
            cv2.imshow(f"Forehead {idx + 1}", cv2.cvtColor(forehead_image, cv2.COLOR_RGB2BGR))  # Display in BGR

        # Save the extracted regions as images if needed
        cv2.imwrite(f"face_{idx + 1}.jpg", cv2.cvtColor(face_image, cv2.COLOR_RGB2BGR))
        cv2.imwrite(f"forehead_{idx + 1}.jpg", cv2.cvtColor(forehead_image, cv2.COLOR_RGB2BGR))
else:
    print("No faces found in the image.")

# Wait for user input to close the images
cv2.waitKey(0)
cv2.destroyAllWindows()
```
